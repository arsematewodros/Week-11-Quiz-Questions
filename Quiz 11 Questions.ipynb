1. Select a random sample of 40 individuals from the population array and assign the created numpy array to the variable sample.

```python
sample = np.random.choice(population_array, size=40, replace=False)
```

2. Calculate the sample mean and assign the value to the variable sample_mean.

```python
sample_mean = np.mean(sample)
```

3. Calculate the standard deviation of the sample and assign the value to the variable sample_std.

```python
sample_std = np.std(sample, ddof=1)
```

4. Calculate the standard error of the sample and assign the value to the variable standard_error.

```python
standard_error = sample_std / np.sqrt(40)
```

5. Calculate the confidence coefficient for a 95% confidence level, given the sample size and degrees of freedom, and assign the value to the variable t_critical.

```python
t_critical = 2.02
```

6. Calculate the lower bound of the confidence interval for the sample mean and assign the result to the variable lower_bound.

```python
lower_bound = sample_mean - t_critical * standard_error
```

7. Calculate the upper bound of the confidence interval for the sample mean and assign the result to the variable upper_bound.

```python
upper_bound = sample_mean + t_critical * standard_error
```

8. Create an array of 1000 bootstrapped sample means (using the sample size of 40) and assign the array to the variable boot_means and sort the values in the array.

```python
boot_means = []
for _ in range(1000):
    boot_sample = np.random.choice(sample, size=40, replace=True)
    boot_means.append(np.mean(boot_sample))
boot_means = np.array(boot_means)
```

9. Show the bootstrapped value representing the 2.5th percentile.

```python
lower_percentile = np.percentile(boot_means, 2.5)
```

10. Show the value bootstrapped representing the 97.5th percentile.

```python
upper_percentile = np.percentile(boot_means, 97.5)
```
